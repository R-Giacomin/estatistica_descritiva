## Frequência em escala intervalar

```python
def faixa(x):
  try:
      x = int(x)
  except:
      return pd.NA

  if x<1: return "0-1"
  elif x<2: return "1-2"
  elif x<3: return "2-3"
  elif x<4: return "3-4"
  elif x<5: return "4-5"
  elif x<6: return "5-6"
  elif x<7: return "6-7"
  elif x<8: return "7-8"
  elif x<9: return "8-9"
  elif x<10: return "9-10"

  return pd.NA

ideb['faixas'] = ideb.IDEB12_17.apply(faixa);
ideb.faixas.value_counts().sort_index()
```

# Gráficos

## Gráfico de barras

```python
# Dados: cores e suas frequências
categorias = ['Vermelho', 'Azul', 'Verde', 'Amarelo']
frequencias = [15, 20, 10, 5]

# Criação do gráfico de barras
plt.figure(figsize=(6,4))
plt.bar(categorias, frequencias, color=['red', 'blue', 'green', 'yellow'])
plt.title("Frequência de Cores Favoritas")
plt.xlabel("Cores")
plt.ylabel("Frequência")
plt.show()
```

## Gráfico de área

```python
# Criação do gráfico de pizza
plt.figure(figsize=(6,6))
plt.pie(frequencias, labels=categorias, autopct='%1.1f%%', startangle=90, colors=['red', 'blue', 'green', 'yellow'])
plt.title("Distribuição Percentual das Cores Favoritas")
plt.show()
```

## Gráfico de pareto

```python
# Ordena os dados de forma decrescente
indices = np.argsort(frequencias)[::-1]
categorias_ordenadas = np.array(categorias)[indices]
frequencias_ordenadas = frequencias[indices]

# Calcula o percentual acumulado
acumulado = np.cumsum(frequencias_ordenadas) / np.sum(frequencias_ordenadas) * 100

fig, ax1 = plt.subplots(figsize=(7,5))

# Gráfico de barras (frequência)
ax1.bar(categorias_ordenadas, frequencias_ordenadas, color=['blue', 'red', 'green', 'yellow'])
ax1.set_xlabel('Cores')
ax1.set_ylabel('Frequência')
ax1.tick_params(axis='y')

# Linha do percentual acumulado
ax2 = ax1.twinx()
ax2.plot(categorias_ordenadas, acumulado, color='#323262', marker='D', ms=7)
ax2.set_ylabel('Percentual Acumulado', color='#323262')
ax2.tick_params(axis='y', labelcolor='#323262')
ax2.set_ylim(0, 110)

plt.title("Gráfico de Pareto das Cores Favoritas")
plt.show()
```

## Gráfico de frequência acumulada

```python
import matplotlib.pyplot as plt
import numpy as np

# Gerando dados aleatórios (por exemplo, distribuição normal)
data = np.random.normal(loc=50, scale=10, size=200)

# Definindo os intervalos (bins)
bins = np.arange(20, 80, 5)  # intervalos de 20 a 75 com amplitude 5

# Calculando a frequência de cada intervalo
hist, bin_edges = np.histogram(data, bins=bins)

# Calculando a frequência acumulada
cumulative = np.cumsum(hist)

# Plotando o gráfico de frequência acumulada (ogiva)
plt.figure(figsize=(8, 5))
plt.plot(bin_edges[1:], cumulative, marker='o', linestyle='-', color='blue')
plt.title('Gráfico de Frequência Acumulada (Ogiva)')
plt.xlabel('Valor')
plt.ylabel('Frequência Acumulada')
plt.grid(True)
plt.show()
```
